```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("d6_2m_5tm.csv")
```

```python
df.head()
```

```python
pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M:%S')
```

```python
for i in range(len(df)):
    if i < 1500:
        df['bathroomLight'][i] = 0
    else:
        df['bathroomLight'][i] = 1
```

```python
for j in range(len(df)):
    if j == 1500:
        df['bathroomDoor'][j] = 1
    else:
        df['bathroomDoor'][j] = 0
```

```python
df_new = df[(df['timestamp']<='2016-02-01 23:59:00') & (df['timestamp']>='2016-02-01 08:00:00')]
```

```python
df['timestamp'].loc[len(df)-1]
```

```python
%matplotlib qt
plt.style.use('seaborn-white')

fig, ax = plt.subplots()
ax.set_title('Badezimmer Aktivität', fontsize=18)
fig.set_size_inches(10, 6.5, forward=True)

plt.ylabel('Badezimmer Licht', fontsize=15)
plt.xlabel('Zeit', fontsize=15)

plt.plot(df_new['timestamp'], df_new['bathroomLight'], 'ro', label='Badezimmer Daten')

ax=plt.gca()

start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 100))
plt.xticks(rotation=25)

plt.grid()
plt.gca().set_ylim(0)

plt.legend(loc="upper left", fontsize=12)
# plt.savefig("sensor.png")
plt.show()
```

```python
%matplotlib qt
plt.style.use('seaborn-white')

fig, ax = plt.subplots()
ax.set_title('Türaktivitäten', fontsize=18)
fig.set_size_inches(10, 6.5, forward=True)

plt.ylabel('die Tür des Badezimmers', fontsize=15)
plt.xlabel('Zeit', fontsize=15)

plt.plot(df_new['timestamp'], df_new['bathroomDoor'], 'bx', label='die Tür Daten')

ax=plt.gca()

start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 100))
plt.xticks(rotation=25)

plt.grid()
plt.gca().set_ylim(0)

plt.legend(loc="upper right", fontsize=12)
# plt.savefig("sensor.png")
plt.show()
```
